```python
%matplotlib inline
import numpy as np
from scipy import integrate
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
from matplotlib import pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
xmin=0
xmax=2*np.pi
sigma=2*np.pi/25.
Nmax=10
nx = 1000
x=np.linspace(xmin,xmax,nx)
```

```python
def u(n,A,B,x,t):
    return A*np.cos(n*(x-t)) + B*np.sin(n*(x-t))
```

```python
def u0(n,x):
    return np.sin(n*x)

def u0_prime(n,x):
    return np.cos(n*x)
```

```python
def u_analytic0(x):
    return np.exp(-((x-np.pi)/sigma)**2)
```

```python
def u_prime_analytic0(x):
    return -(2.0/sigma**2)*(x-np.pi)*np.exp(-((x-np.pi)/sigma)**2)
```

```python
def inner(foo,bar):
    return integrate.quad(lambda x: (1/np.pi)*foo(x)*bar(x),xmin,xmax)[0]
```

```python
Bcoeffs = [inner(lambda x: u0(i+1,x),u_analytic0) for i in range(Nmax)]
Acoeffs = [inner(lambda x: u0_prime(i+1,x),u_analytic0) for i in range(Nmax)]
```

```python
def uf(x,t):
    return sum([u(i+1,Acoeffs[i],Bcoeffs[i],x,t) for i in range(len(Bcoeffs))])

def udiff(rho):
    return sum(np.array([Acoeffs[i]*np.cos((i+1)*rho) + Bcoeffs[i]*np.sin((i+1)*rho) for i in range(len(Acoeffs))]))
```

```python
text_x=0.1
text_y=1.
text = "Waves = {}".format(Nmax)
fig = plt.figure()
ax = plt.axes(xlim=(0,2*np.pi),ylim=(-0.25,1.1))
label = ax.text(text_x,text_y,text)
line, = ax.plot([],[],lw=4)
dt = 0.05

def init():
    line.set_data([],[])
    return line

def animate(i):
    t = dt*i
    y = udiff(x-t)
    line.set_data(x,y)
    return line,

nf = int(np.ceil(2*np.pi/dt))
anim = animation.FuncAnimation(fig,animate,init_func=init,frames=nf,interval=int(nf/10),blit=True)
anim.save('ring_cavity_n{}.gif'.format(Nmax),writer='imagemagick')
anim
```
